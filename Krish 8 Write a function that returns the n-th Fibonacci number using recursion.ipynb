#### Fibonacci Recursive

```python
### My Solution, even though I was in doubt, I wrote correct hihi
def fib_recur(n):
        if n == 0:
                return 0
        elif n == 1:
                return 1
        else:
                return fib_recur(n-1)  +  fib_recur(n-2)  

fib_recur(8)
```

```python
#### krish naik
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

fibonacci(10)
```
